# FOR 3 NODAL POINTS

```python
import numpy as np
import matplotlib.pyplot as plt
import math
x=np.linspace(0,1,3) #3 nodal points
xh=np.linspace(0,1,1000)
h=x[1]-x[0]
f,f2=[],[]
x1=[]
m=0.6281834549054398
for i in range(len(x)):
    f.append(m*x[i])
    f2.append(m*x[i])
n=0
while n<500:
    for i in range(1,len(x)-1):
        f2[i]=(0.5*(f[i+1]+f[i-1]))/(1+h**2)
    for i in range(len(x)):
        f[i]=f2[i]
    n=n+1
y1=[]
for i in xh:
    y1.append((math.exp(math.sqrt(2)*i)-math.exp(-1*math.sqrt(2)*i))/(math.sqrt(2)*(math.exp(math.sqrt(2))+math.exp(-1*math.sqrt(2)))))
plt.plot(x,f,'bo')
plt.plot(xh,y1)
```
